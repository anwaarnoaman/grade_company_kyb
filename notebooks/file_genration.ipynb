```python
from fpdf import FPDF
import os

# Directory to save PDFs
output_dir = "synthetic_documents"
os.makedirs(output_dir, exist_ok=True)

# Document templates
documents = [
    {
        "filename": "trade_license.pdf",
        "title": "Trade License",
        "content": """TRADE LICENSE
Company Name: Acme FinTech Solutions LLC
License Number: TL-2026-987654
Issuing Authority: Dubai Department of Economic Development
Legal Form: Limited Liability Company
Issue Date: 01-Jan-2024
Expiry Date: 31-Dec-2026
Authorized Signatory: John Smith, CEO"""
    },
    {
        "filename": "moa_aoa.pdf",
        "title": "Memorandum of Association",
        "content": """MEMORANDUM OF ASSOCIATION
Company Name: Acme FinTech Solutions LLC
Registered Office: Dubai Silicon Oasis, Dubai, UAE
Legal Form: Limited Liability Company

OBJECTS:
1. To provide financial technology services.
2. To operate digital payment platforms.

SHARE CAPITAL:
Total Shares: 1000
Par Value: AED 1

SHAREHOLDERS:
- Alice Johnson: 40%
- Bob Lee: 30%
- Charlie Wong: 30%

BOARD OF DIRECTORS:
- John Smith (CEO)
- Mary Brown (CFO)"""
    },
    {
        "filename": "board_resolution.pdf",
        "title": "Board Resolution",
        "content": """BOARD RESOLUTION
Acme FinTech Solutions LLC
Date: 15-Feb-2026

Resolved that:
Mr. John Smith, CEO, is authorized to open a corporate bank account with Future Bank UAE and sign all related documents.

Certified by:
Mary Brown, Company Secretary"""
    },
    {
        "filename": "id_john_smith.pdf",
        "title": "Authorized Signatory ID",
        "content": """UAE Passport
Name: John Smith
Passport No: P12345678
Nationality: UAE
Date of Issue: 01-Jan-2022
Date of Expiry: 01-Jan-2032"""
    },
    {
        "filename": "bank_letter.pdf",
        "title": "Bank Letter",
        "content": """Future Bank UAE
Corporate Banking Division

To Whom It May Concern:

This is to certify that Acme FinTech Solutions LLC maintains a corporate account with us.
Account Number: 9876543210
Account Status: Active
Average Monthly Balance: AED 2,500,000

Issued on: 20-Feb-2026
Authorized Signatory: Sarah Al Farsi, Bank Manager"""
    },
    {
        "filename": "vat_certificate.pdf",
        "title": "VAT / TRN Certificate",
        "content": """VAT Registration Certificate
Company Name: Acme FinTech Solutions LLC
TRN: 100123456789
Jurisdiction: UAE
Registration Date: 01-Jan-2024
Status: Active"""
    },
    {
        "filename": "balance_sheet.pdf",
        "title": "Balance Sheet",
        "content": """Acme FinTech Solutions LLC
Balance Sheet as of 31-Dec-2025

ASSETS
Cash & Bank: 3,000,000
Accounts Receivable: 500,000
Total Assets: 3,500,000

LIABILITIES
Accounts Payable: 800,000
Loans: 1,200,000
Total Liabilities: 2,000,000

EQUITY
Share Capital: 1,000,000
Retained Earnings: 500,000
Total Equity: 1,500,000"""
    },
    {
        "filename": "profit_loss.pdf",
        "title": "Profit & Loss Statement",
        "content": """Acme FinTech Solutions LLC
Profit & Loss Statement for FY 2025

Revenue: 5,000,000
Cost of Goods Sold: 2,500,000
Gross Profit: 2,500,000
Operating Expenses: 1,000,000
Net Profit: 1,500,000
Audit Status: Audited"""
    }
]

# Function to create PDF
def create_pdf(doc):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", "B", 16)
    pdf.cell(0, 10, doc["title"], ln=True, align="C")
    pdf.ln(10)
    pdf.set_font("Arial", "", 12)
    for line in doc["content"].split("\n"):
        pdf.multi_cell(0, 8, line)
    pdf.output(os.path.join(output_dir, doc["filename"]))

# Generate all PDFs
for doc in documents:
    create_pdf(doc)

print(f"All synthetic KYB documents generated in folder '{output_dir}'")
```
